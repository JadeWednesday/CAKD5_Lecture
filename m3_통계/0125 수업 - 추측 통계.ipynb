## 확률의 수학적 정의와 의미
<br>

####  확률표본
    확률표본 또는 표본은 풀고자 하는 문제에서 발생할 수 있는 하나의 현상,
    혹은 선택될 수 있는 하나의 경우를 말한다.
    
#### 표본공간
    가능한 모든 표본의 집합을 말하며,
    고려하는 범위에서 어떤 표본(경우, 현상)이 가능하고
    또 어떤 표본이 가능하지 않은지 정의하는 작업

#### 사건        
    표본공간의 부분집합, 즉 전체 표본공간 중에서 관심을 가지는 일부 표본의 집합
<br>

#### 확률
    - 사건(부분집합)을 입력하면 숫자(확률값)가 출력되는 함수
    - 모든 각각의 사건(부분집합)에 어떤 숫자를 할당하는 함수
    - 확률은 표본이 아닌 사건을 입력을 가지는 함수
<br><br>

#### 확률의 의미
- 빈도주의 관점: 반복적으로 선택된 표본이 사건 A의 원소가 될 경향.<br>
    동전을 던져 앞면이 나오는 사건의 확률값은 0.5
- 베이지안 관점: 선택된 표본이 특정한 사건에 속한다는 가설, 명제 혹은 주장의 신뢰도.<br>
    반복이라는 개념은 사용되지 않는다.<br>
    새가 날 수 있는 가능성은 95%, '앞면이 나왔다'는 주장의 신뢰도는 50%
    * 베이지안 확률론에서 사건이 일어났다, 혹은 발생했다는 말은<br>
        그 사건의 원소 중에 정말로 선택된 표본이 있다는 사실을 알게 되었다는 것을 뜻한다.
<br><br>

#### 확률의 성질
- 공집합의 확률: 공집합인 사건의 합은 0이다.
    P(0) = 0
- 여집합의 확률: 어떤 사건의 여집합인 사건의 확률은 (1-원래 사건의 확률)과 같다.
- 포함-배제의 원리:
    두 사건의 합집합의 확률은 (각 사건의 확률의 합 - 두 사건의 교집합의 확률)과 같다.
- 전체 확률의 법칙: '생일이 1월인 남자'라는 사건의 확률과 '생일이 2월인 남자'라는 사건의 확률,<br>
    '이외에도 각각의 월을 생일로 가진 남자'라는 사건들의 확률을 모두 합하면<br>
    '남자'라는 사건의 확률이 된다.<br>
    각각의 월에 있는 남자의 생일에 서로 공통 원소가 없기 때문이다. (남자라는 요소를 제외하면)
<br><br>

#### 확률분포함수
- 확률이 어디에 어느 정도 분포되어 있는지를 수학적으로 명시하고 전달하는 도구.
- 만약 표본의 개수가 무한하다면, 현실적으로 모든 사건을 하나하나 기술하는 것은 불가능하며,<br>
    확률분포함수를 이용하여 이 문제를 해결할 수 있다.
    * 확률질량함수: 유한 개의 사건이 존재하는 경우, 각 단수 사건에 대한 확률만 정의하는 함수.
        (이산형 데이터)
    * 누적분포함수: 시작점을 모두 똑같이 음의 무한대로 통일한,<br>
        특수한 구간을 사용하는 사건의 확률분포를 묘사하는 함수
    * 확률밀도함수: 누적분포함수를 미분하여 구한 도함수를 의미<br>
        특정한 구간의 확률이 다른 구간에 비해 상대적으로 얼마나 높은가를 나타내는 것이나,<br>
        그 값 자체가 확률은 아니다.
        (연속형 데이터)
<br><br>

#### 확률변수
- 확률적인 숫자값을 출력하는 변수

#### 추측통계
    확률을 사용하여 모집단의 통계적 성질을 추출하는 기법

[모집단 분포와 표본분포]
- 모집단: 어떤 통계적 실험의 대상이 되는 모든 대상물<br>
    ex) 대한민국 모든 남성의 키 데이터, 2018년도 수능 점수 데이터
- 모집단 분포: 모집단은 데이터 집단이므로 어떠한 분포를 이루며,<br>
    이 데이터가 이루는 확률 분표가 모집단 분표
- 모수: 모집단의 특성을 나타내는 수치<br>모평균, 모분산, 모표준편차, 모비율, 모상관관계 등
- 표본: 모집단의 모수(특성)를 알기 위하여 표본을 추출하여 모집단의 모수를 추정
- 확률표본: 모집단을 구성하는 모든 구성원이 표본으로 뽑힐 확률을 동일하게 하여 추출한 표본<br>
    'Random'이란 무작위라는 의미로 어떠한 의도를 가지지 않는다는 것을 의미하며,<br>
    모든 구성원이 뽑힐 확률이 동일하다는 개념이다.
    * 표본에 편향(bias)이 생기면 모수를 추측하기 어려워지기 때문에 불편성을 지녀야 함
- 통계량: 확률표본의 특성을 설명하는 값<br>표본평균, 표본분산, 표본표준편차, 표본비율, 표본상관관계 등이 포함된다.<br>
    모수는 모집단이 변하지 않기 때문에 그 값이 불변이지만,<br>
    통계량은 표본을 어떻게 추출하느냐에 따라서 그 값이 다르게 나타난다.
- 확률변수: 무작위 실험을 했을 때 특정 확률로 발생하는 각각의 결과를 수치값으로 표현한 변수.<br>ex) 동전의 앞면과 뒷면이 나올 각각의 확률
- 확률분포: 확률변수 X의 모든 값과 그에 대응하는 확률값들의 분포
- 통계량이 확률변수라는 말의 의미는?<br>
    확률변수의 확률분포가 수학적으로 설명 가능한 분포(ex. 정규분포)를 따른다고 할 때,<br>
    특정 확률변수와 확률함수를 알고 있다면 특정 사건이 일어날 확률을 계산할 수 있으며<br>
    더 나아가 모수와 통계량의 관계를 알고 있을 때 그것을 통해 모수에 대한 통계적인 추정을 할 수 있다.<br>
    즉, 통계량의 확률함수 또는 표본분포를 안다고 가정하면 이를 통해 모수를 추정할 수 있다.
    * 만약 모수가 표준정규분포를 따른다면 표준편차 1, 평균이 0일 테니<br>이런 것들을 이용해 모수를 추정할 수 있다는 말임.
<br>

[표본통계량 성질]
- 샘플 사이즈가 크면 표본평균은 모평균에 가까워진다.
- 샘플 사이즈가 커질수록 표본평균의 표준편차가 작아진다.<br>샘플 사이즈를 크게 하면 흩어짐(산포도)이 적은, 신뢰할 수 있는 표본평균을 얻을 수 있다.
- 표준오차는 이론상 표본평균에 대한 표준편차인데, 샘플 사이즈가 커지면 표준오차는 작아진다.
    * 그 이유는, 표본 사이즈가 커지면서 표본평균이 모평균이 가까워져서 그 차이가 작아지기 때문임
- 예를 들어, 모분산은 0.64, 표본분산의 평균값은 0.575로 분산이 과소평가되는 경향이 발생<br>-> 불편분산을 사용해 편향을 제거(ddof=1)
- 샘플 사이즈가 커질수록 불편분산은 모분산에 가까워진다.
- 불편성: 추정량의 기댓값이 진짜 모수가 되는 특성. 편향이 없는 추정량
    - 표본평균과 불편분산은 둘 다 불편성을 가지고 있다고 할 수 있다.
- 일치성: 샘플 사이즈가 커지면 추정량이 진짜 모수에 가까워지는 특성.
    - 샘플 사이즈가 커지면 표본평균은 모평균에,<br>불편분산은 모분산에 가까워지므로 일치성을 가진다.
- 큰 수의 법칙: 표본의 크기가 커지면 표본평균이 모평균에 가까워지는 방법을 표현한 법칙
- 중심극한정리: 모집단의 분포가 어떻든지 간에 샘플 사이즈가 커지면<br>표본평균들이 이루는 분포가 모집단의 평균 $\mu$, 표준편차 $\sigma/sqrt(n)$인 정규분포에 가까워진다.

```python
# 모집단과 표본
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/ch4_scores400.csv')
print(df.shape)
scores = np.array(df.score) # 관측치 400개
scores[:10]
```

```python
# 복원 추출 - 하나 뽑고 뽑은 자리 다시 채워놓음(복원)
# 중복을 허용한다!
np.random.choice([1,2,3],3) # replace=True
```

```python
# 비복원 추출 - 하나 뽑고 뽑은 자리 비워놓음(비복원)
# 중복을 허용하지 않는다
np.random.choice([1,2,3],3,replace=False)
```

```python
# 표본평균 - 20개만 뽑구서
np.random.seed(0)
sample = np.random.choice(scores,20)
sample.mean()
```

```python
# 확률분포
dice = [1,2,3,4,5,6] # 이건 확률변수
prob = [1/21,2/21,3/21,4/21,5/21,6/21] # 주사위이긴 주사위인데 균등분포가 아닌 주사위
np.random.choice(dice, p=prob)
```

```python
np.random.seed(0)
sample = np.random.choice(dice, 100, p=prob) # 100번을 시도해 본다면?
sample
```

```python
freq, _ = np.histogram(sample, bins=6, range=(1,7))
print(freq)
rdf = pd.DataFrame({'frequency': freq,
                   'relative frequency': freq/100},
                  index = pd.Index(np.arange(1,7), name='dice'))
rdf
```

```python
# 복원 추출이기 때문에 무한정 많이 뽑아볼 수 있다.
# 표본의 크기를 크게 늘리면 실제 지정했던 확률분포와 같아지는지 볼까?
sample = np.random.choice(dice, 10000, p=prob)

plt.style.use('seaborn')
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=6, range=(1,7), density=True, rwidth=0.8) # 상대도수는 확률과 비슷한 성질을 지니게 된다.
# 실제 확률 분포를 가로선으로 표시해 보자
ax.hlines(prob, np.arange(1,7),np.arange(2,8),color='crimson') # y, xmin, xmax를 입력(선의 범위)
# 막대그래프
ax.set_xticks(np.linspace(1.5,6.5,6)) # 눈금의 위치는 중앙에 두자
ax.set_xticklabels(np.arange(1,7)) # 눈금의 값은 이걸로 바꿔주세요
ax.set_xlabel('frequency')
ax.set_ylabel('relative frequency')
plt.show()
plt.close()
# 시행횟수를 계속 늘리면 상대도수(상대적인 확률)는 실제의 확률분포로 수렴되어 간다.
```

```python
# [연습문제] 무작위 추출로 표본 크기가 20인 표본을 추출하여 표본평균을 계산하는 작업을
# 10000번 수행하고, 그 결과를 히스토그램으로 그린 후
# 모평균을 중심으로 분포하는지의 여부를 확인하고 모평균 추정의 근거를 설명하세요.
# 모평균을 세로선으로 표시하기

df = pd.read_csv('./data/ch4_scores400.csv')
scores = np.array(df.score)

li = [np.random.choice(scores,20).mean()
     for _ in range(10000)]
    
plt.figure(figsize=(10,8))
plt.hist(li, bins=100, range=(0,100), density=True) # 상대도수 on
plt.vlines(scores.mean(),0,0.13, color='crimson')
plt.axis([50, 90, 0, 0.13]) # xmin, xmax, ymin, ymax
plt.xlabel('Score')
plt.ylabel('Relative Frequency')
plt.show()
```

```python
# [과제] 평균이 4, 표준편차가 0.8인 정규분포를 따르는 모집단으로부터
# 10000개의 표본을 추출하여 저장하고 그 중 10개를 출력하세요.

# scipy.stats.norm()은 가우시안정규분포 객체를 생성하는 메소드
from scipy.stats import norm
pop = norm(loc=4, scale=0.8) # 평균이 4, 표준편차 0.8인 정규분포 객체
np.random.seed(1) # numpy 시드 고정은 scipy에서도 먹히는 듯
sample = pop.rvs(10000) # 정규분포 객체에서 난수를 생성하는 rvs 메소드
sample[:10]
```

```python
# [과제] 위 과제에서 10000개의 표본의 평균 및 표준편차를 구하고
# 시각화하여 정규분포를 따르는지 확인하세요.
print(f'''무작위 표본평균: {sample.mean().round(2)}
무작위 표본 표준편차: {sample.std().round(2)}''')

plt.figure(figsize=(8,6))
plt.hist(sample,bins=100, range=(0,8))
plt.vlines(4, 0, 500, color='crimson',alpha=0.8) # 모평균 표시
plt.show()
sns.histplot(sample,bins=100, kde=True) # sns 버전, 확률밀도함수 on
plt.show()
```

```python
# [과제] 샘플 사이즈를 10000개에서 100000개까지 순차적으로 늘렸을 때
# 표본평균이 모평균에 가까워짐을 시각화하여 입증하세요.

# 표본 1만개씩 늘려가며 평균 낸 걸로 시각화하면 될 듯
li=[]
np.random.seed(0)
for i in range(10000,100001,10000):
    li.append(pop.rvs(i).mean().round(4))

plt.figure(figsize=(8,6))
plt.hlines(4,0,11,color='crimson',alpha=0.8) # 모평균 표시
plt.bar(np.arange(1,11),li,width=0.8)
plt.xticks(np.arange(1,11),np.arange(10000,100001,10000))
plt.ylim(3.98, 4.01)
plt.xlabel('Sampling Number')
plt.ylabel('Mean')
plt.show()
np.array(li)
```

```python
# 조남현님 풀이
# 라인 플롯을 이용해서 표본평균이 모평균에 수렴하는 모양을 나타냄
from scipy.stats import norm
plt.style.use('seaborn')

pop = norm(loc=4,scale=0.8)
arr = np.arange(10000,100001,10) # 표본 크기 10단위로
arr_mean = np.empty(len(arr))
for i in range(0,len(arr)):
    species = pop.rvs(size=arr[i]) # 표본 크기 배열에서 뽑아서
    arr_mean[i] = np.mean(species) # 그 크기 샘플의 평균을 모은 배열 만들기

fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)
ax.plot(arr,arr_mean) # 표본의 크기 arr, 그 크기일 때의 평균 arr_mean
ax.set_ylabel('mean of species', fontsize = 20)
ax.set_xlabel('number of species',fontsize = 20)
ax.set_title('Central Limit Theorem',fontsize = 50)
ax.set_xticks(np.linspace(10000,100000,10))
ax.axhline(y=pop.mean(),color = 'red') # 모평균
plt.show()
```

```python
# 선생님 풀이
size_array = np.arange(10,100100,100)
population = norm(loc=4, scale=0.8)

# 샘플 사이즈를 변화시키면서 표본평균을 도출하는 시뮬레이션
np.random.seed(1)
sample_mean_array = np.zeros(len(size_array))
for i in range(0,len(size_array)):
    sample = population.rvs(size_array[i]) # i가 0일 때 표본 10개, 1일 때 110개, ...
    sample_mean_array[i] = np.mean(sample)

# 샘플 사이즈가 커질수록 표본평균이 모평균에 수렴한다.

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax.plot(size_array, sample_mean_array)
ax.hlines(4, 0,100100, color='crimson', alpha=0.8)
plt.xlabel('Sample Size')
plt.ylabel('Sample Mean')
ax.set_xticks(np.linspace(0,100000,11))
plt.show()
```

```python
# [과제] 샘플 사이즈가 커질수록 표본평균의 표준편차가
# 모집단의 표준편차로 수렴한다는 것(표본의 표준편차가 작아진다)을 시각화하여 증명하세요.
# y축이 표본평균의 표준편차

rv = norm(loc=4, scale=0.8)
sizearr = np.arange(100,100100,100)
mean = np.zeros(len(sizearr))
std = np.zeros(len(sizearr))
for i in range(0,len(sizearr)):
    sample = rv.rvs(sizearr[i]) # 표본 만들기
    std[i] = rv.std()/(len(sample)**0.5)

plt.figure(figsize=(6,4))
plt.plot(sizearr,std)
plt.show()
```

```python
# 어정호님 풀이
import numpy as np
from scipy import stats

size_array  = np.arange(10,100100, 100)
len(size_array)
population = stats.norm(loc=4,scale=0.8)
population.rvs(size=10)
# 샘플사이즈를 변화시키면서 표본평균을 도출하는 시뮬레이션
np.random.seed(1)
sample_mean_array_size = np.zeros(len(size_array))
aaa = np.zeros(len(size_array))
for i in range(0,len(size_array)):
    sample = population.rvs(size=size_array[i]) # 표본 크기
    sample_mean_array_size[i] = np.mean(sample) # 표본 평균
    aaa[i]= population.std() / (len(sample)**0.5) # 표본 표준편차
    # 모집단 표준편차를 샘플크기 루트로 나누면 표본의 표준편차가 나온다.. 고 함
sample_mean_array_size[:10]

# 샘플사이즈가 커질수록 표본평균이 모평균에 가까워짐
import matplotlib.pyplot as plt
plt.plot(size_array, aaa, color='black')
plt.xlabel('sample size')
plt.ylabel('sample std')
plt.show()
```

```python
# [과제] 샘플 사이즈가 커질수록 표본평균의 흩어지는 정도(산포도)가 작아져서
# 모평균에 가깝게 모이는 현상을 바이올린 플롯으로 시각화하여 출력하세요.
# 바이올린 플롯을 사용하면 샘플의 분포도를 보여준다.
# 샘플은 다양한 구간에서 3개 정도 추출하기(위에서 만든 거 써도 되고)

from scipy.stats import norm
np.random.seed(0)
pop = norm(loc=5, scale=0.7) # 평균 5, 표준편차 0.6
mean = pd.DataFrame({'Mean':[pop.rvs(100).mean(),
                         pop.rvs(1000).mean(),    
                         pop.rvs(5000).mean(),
                         pop.rvs(10000).mean(),
                         pop.rvs(100000).mean()]},
                index=pd.Index(['100개','1000개','5000개','10000개','100000개'],name='표본 개수'))
display(mean)
sns.violinplot(data=mean,color='mediumaquamarine')
plt.hlines(5,-.5,.5,color='crimson')
plt.xlabel('Sampling')
plt.ylabel('Mean')
plt.show()
```
