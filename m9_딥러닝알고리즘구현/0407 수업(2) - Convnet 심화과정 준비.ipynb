[과제] 캐글에서 트레인 폴더 다운 받아서 강아지와 고양이 데이터 셋 만들기

```python
import os, shutil
from google.colab import drive
drive.mount('/content/gdrive')

origin = '/content/gdrive/MyDrive/cakd5_colab/m9_딥러닝알고리즘구현/cats_and_dogs/train'
base = '/content/gdrive/MyDrive/cakd5_colab/m9_딥러닝알고리즘구현/cats_and_dogs/cad'
```

```python
# 학습/검증/테스트 용 폴더 생성

os.mkdir(base)
train_dir = os.path.join(base, 'train')
os.mkdir(train_dir)
val_dir = os.path.join(base, 'validation')
os.mkdir(val_dir)
test_dir = os.path.join(base, 'test')
os.mkdir(test_dir)
```

```python
# 각 폴더에 고양이/강아지 하위 폴더 생성

train_cat = os.path.join(train_dir, 'cats')
os.mkdir(train_cat)
val_cat = os.path.join(val_dir, 'cats')
os.mkdir(val_cat)
test_cat = os.path.join(test_dir, 'cats')
os.mkdir(test_cat)

train_dog = os.path.join(train_dir, 'dogs')
os.mkdir(train_dog)
val_dog = os.path.join(val_dir, 'dogs')
os.mkdir(val_dog)
test_dog = os.path.join(test_dir, 'dogs')
os.mkdir(test_dog)
```

```python
# 고양이 사진 1000, 500, 500개로 나눠 각각 학습, 검증, 테스트 폴더에 나눠 넣기

fname = ['cat.{}.jpg'.format(i) for i in range(1000)]
for f in fname:
  source = os.path.join(origin, f)
  copy = os.path.join(train_cat, f) # 학습
  shutil.copyfile(source, copy)

fname = ['cat.{}.jpg'.format(i) for i in range(1000,1500)]
for f in fname:
  source = os.path.join(origin, f)
  copy = os.path.join(val_cat, f) # 검증
  shutil.copyfile(source, copy)

fname = ['cat.{}.jpg'.format(i) for i in range(1500,2000)]
for f in fname:
  source = os.path.join(origin, f)
  copy = os.path.join(test_cat, f) # 테스트
  shutil.copyfile(source, copy)
```

```python
# 강아지 사진 1000, 500, 500개로 나눠 각각 학습, 검증, 테스트 폴더에 나눠 넣기

fname = ['dog.{}.jpg'.format(i) for i in range(1000)]
for f in fname:
  source = os.path.join(origin, f)
  copy = os.path.join(train_dog, f) # 학습
  shutil.copyfile(source, copy)

fname = ['dog.{}.jpg'.format(i) for i in range(1000,1500)]
for f in fname:
  source = os.path.join(origin, f)
  copy = os.path.join(val_dog, f) # 검증
  shutil.copyfile(source, copy)

fname = ['dog.{}.jpg'.format(i) for i in range(1500,2000)]
for f in fname:
  source = os.path.join(origin, f)
  copy = os.path.join(test_dog, f) # 테스트
  shutil.copyfile(source, copy)
```

```python
print(f'학습용 고양이 사진 >> {len(os.listdir(train_cat))}장')
print(f'검증용 고양이 사진 >> {len(os.listdir(val_cat))}장')
print(f'테스트용 고양이 사진 >> {len(os.listdir(test_cat))}장\n')
print(f'학습용 강아지 사진 >> {len(os.listdir(train_dog))}장')
print(f'검증용 강아지 사진 >> {len(os.listdir(val_dog))}장')
print(f'테스트용 강아지 사진 >> {len(os.listdir(test_dog))}장')
```
