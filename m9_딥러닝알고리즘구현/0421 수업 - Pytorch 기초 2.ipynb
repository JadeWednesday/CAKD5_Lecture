```python
import torch
```

```python
# Manipulations
# 인덱싱: numpy처럼 인덱싱 형태로 사용 가능

x = torch.Tensor([[1,2],[3,4]])
print(x)

print(x[0,0])
print(x[:,0])
print(x[0,:])
```

```python
# view : 텐서의 크기(size)나 모양(shape)을 변경
x = torch.randn(4,5)
print(x.shape)
print(x.size())
```

```python
print(x,'\n')
y = x.view(20)
print(y)
print(y.shape)
```

```python
y = x.view((2,2,5))
print(y)
print(y.shape)
```

```python
# [과제]
x = torch.FloatTensor([[1,2],[3,4],[5,6],[7,8]])
print(x)

# size를 활용한 텐서 사이즈 보기
print(x.size())

# shape를 활용한 텐서 사이즈 보기
print(x.shape)

# dim을 활용한 텐서 차원 확인
print(x.dim())

# 0 차원 사이즈 확인
print(x.size(0))

# 1 차원 사이즈 확인
print(x.shape[1])
```

```python
# 차이를 조금 익혀볼까?
import numpy as np

print(np.array([1,2,3]).shape, np.array([1,2,3]).ndim)
print(np.array([[1,2,3]]).shape, np.array([[1,2,3]]).ndim)
print(np.array([[1],[2],[3]]).shape, np.array([[1],[2],[3]]).ndim)
```

```python
# item(): 해당 값을 자르지 않고 반환
x = torch.randn(1)
print(x)
print(x.item())
print(x.dtype)
```

```python
# 스칼라값 하나만 존재할 때에만 item() 사용 가능
x = torch.randn(2)
print(x)
# print(x.item())
print(x.dtype)
```

```python
# squeeze : 차원을 축소
tensor = torch.rand(1,3,3)
print(tensor)
print(tensor.shape,'\n')
t = tensor.squeeze()
print(t)
print(t.shape) # 아마도 1개 짜리는 날려도 텐서의 모양 자체는 변하지 않으니 1 값인 차원이 있으면 그걸 날려버리는 기믹인 듯
```

```python
# unsqueeze : 차원을 증가
t = torch.rand(3,3)
print(t)
print(t.shape,'\n')
tensor = t.unsqueeze(dim=0) # 0번째, 즉 3차원 방향으로 차원을 증가시켰다.
print(tensor)
print(tensor.shape)
```

```python
tensor = t.unsqueeze(dim=2)
print(tensor)
print(tensor.shape) # 열 방향으로 차원 늘리기
```

```python
tensor = t.unsqueeze(dim=1)
print(tensor)
print(tensor.shape) # 행 방향으로 차원 늘리기
```
