```python
# [과제] df의 Age 컬럼을 이용하여 새로운 컬럼 Age_cat을 람다함수를 이용하여 출력하세요.
# 카테고리 7개
# 사용자 함수 만들고 그걸 람다 함수로 매핑하면 될 듯

def cat(age):
    c = ''
    if age<10: c = 'Child'
    elif age <20: c = 'Teen'
    elif age <30: c = '20s'
    elif age <40: c = '30s'
    elif age <50: c = '40s'
    elif age <60: c = '50s'
    else: c='Eldery'
    return c
df['Age_cat'] = df.Age.apply(lambda x:cat(x))
df[['Age', 'Age_cat']].head()
```

```python
# 호칭추출. 뭐가 있나 먼저 확인
dfa = df.copy()
name_grade = dfa['Name'].apply(lambda x : x.split(", ",1)[1].split(".")[0])
name_grade = name_grade.unique().tolist()
print(name_grade)

# 호칭에 따라 사회적 지위(1910년대 기준)를 정의합니다.
grade_dict = {'A': ['Rev', 'Col', 'Major', 'Dr', 'Capt', 'Sir'], # 명예직
              'B': ['Ms', 'Mme', 'Mrs', 'Dona'],                 # 여성
              'C': ['Jonkheer', 'the Countess'],                 # 귀족이나 작위
              'D': ['Mr', 'Don'],                                # 남성
              'E': ['Master'],                                   # 어린남성
              'F': ['Miss', 'Mlle', 'Lady']}                     # 젊은 여성

# 이름 재정의 사용자 함수
def give_grade(x):
    grade = x.split(", ", 1)[1].split(".")[0]
    for key, value in grade_dict.items():
        for title in value:
            if grade == title:
                return key

# Name에 적용 및 확인
dfa['Name'] = dfa['Name'].apply(lambda x: give_grade(x))
print(dfa['Name'].value_counts())

# 사회적 지위에 따라 보통 나이가 비슷. 지위를 참고해 나이 결측치 대체
import warnings
warnings.filterwarnings('ignore')

for i in dfa.Name:
    dfa.Age[(dfa.Name == i)&(dfa.Age.isnull())] = dfa.Age[dfa.Name == i].mean()
dfa.info()
```

```python
# [과제] t_df에 대하여 결손값 처리, 불필요한 컬럼 삭제, 인코딩을 수행하는 전처리 과정을
# 하나의 사용자 함수(def transform_features(df))로 작성하여 분석용 데이터셋을 만드세요.

dfa = df.copy()

def transfrom_features(t):
    t1 = t.copy()
    t1['Family'] = t.SibSp + t.Parch
    t1.drop(['PassengerId','Name','Ticket','Name_len','Age_cat','SibSp','Parch'],axis=1,inplace=True) # 사용하지 않을 열 삭제
    for i in t1.columns:
        if t1[i].isna().sum() != 0:                         # 결측값이 있는 것들
            if t1[i].isna().sum().sum() > len(t) * 0.7:  # 결측값 개수가 데이터 개수의 70프로를 넘는 경우 그 열 안씀
                t1.drop(i,axis=1,inplace=True)
            elif t1[i].dtype == 'O':                        # 문자인 경우 최빈값으로 대체
                many = t1[i].value_counts().index[0]
                t1[i] = t1[i].fillna(many)
            else:
                t1[i][t1[i].isna()] = t1[i].mean()          # 숫자인 경우 평균값으로 대체
    return pd.get_dummies(t1)
transfrom_features(dfa)
```
