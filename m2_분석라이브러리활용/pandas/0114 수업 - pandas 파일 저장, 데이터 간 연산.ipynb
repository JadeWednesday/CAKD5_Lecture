### 연산
- 행/열 인덱스를 기준으로 모든 원소를 정렬한다.
- 동일한 위치에 있는 원소끼리 일대일 대응시킨다.
- 일대일 대응이 되는 원소끼리 연산을 처리한다.

```python
st1 = pd.Series({'국어':98,'영어':95,'수학':54})
st2 = pd.Series({'수학':87,'국어':92,'영어':97})

add = st1+st2
sub = st1-st2
mul = st1*st2
div = round((st1/st2),2)

print(type(add))
df = pd.concat([add,sub,mul,div],axis=1)
df.columns = ['덧셈','뺄셈','곱셈','나눗셈']
df # 인덱스 위치가 제멋대로여도 알아서 정렬해서 연산해 줌
```

```python
result = pd.DataFrame([add,sub,mul,div],index=['덧셈','뺄셈','곱셈','나눗셈'])
result
# 데이터프레임화 시키면 컨캣으로는 못 넣었던 인덱스 넣기도 가능해짐.
# 이때엔 딕셔너리 타입의 키값이 칼럼으로 들어간다
```

```python
df1 = pd.DataFrame(np.arange(12.).reshape(3,4),columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape(4,5),columns=list('abcde'))

df2.loc[1,'b'] = np.nan

add = df1+df2 # df1.add(df2,fill_value="뿅")
sub = df1-df2
mul = df1*df2
div = round((df1/df2),2)
display(add,sub,mul,div)
```

```python
display(df1.add(df2,fill_value=0),df1.sub(df2,fill_value=0),
        df1.mul(df2,fill_value=0),df1.div(df2,fill_value=0).round(2))
```
