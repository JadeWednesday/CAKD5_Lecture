```python
# [연습문제] 전체 행을 2행 간격으로 선택, 역순으로 인덱싱
data = pd.read_csv('./dataset/file_data.csv')
data['가격'] = [7000,6500,5500,5500,4500]
data.loc[5] = ['떡볶이',3000]

df = pd.DataFrame({
    '차림표':['비빔밥','돈까스','라면','제육덮밥','알밥'],
    '가격': [7000,8500,4500,5500,7500]})
# 두 데이터를 합치는 것부터 시작

d = pd.concat([data,df],axis=1)
# 2행 간격으로 선택
display(d.loc[0:5:2],d.loc[[0,2,4]])
# 역순 인덱싱
display(d.loc[::-1])
```

```python
# [과제] iloc, loc를 이용해서 아래 요구사항들을 출력하세요.
# 의 kor열을 series와 dataframe으로 각각 출력
# df에서 Sam의 math 점수를 출력 (2가지 방법)
# df에서 Tom의 eng와 sci를 출력 (2가지 방법)
# df에서 Jim, Tim의 eng, gym 점수를 출력
# df에서 Jim, Sam의 art, math, gym 점수를 출력

columns = ['kor','eng','art','math','gym','sci']
index = ['Jim','Sam','Tom','Tim','Bob']
np.random.seed(0)
data = np.random.randint(70,101,size=(5,6))
df = pd.DataFrame(data,index,columns)
df
```

```python
# [과제] iloc, loc를 이용해서 아래 요구사항들을 출력하세요.
# df에서 Sam의 math 점수를 100으로 변경
# df에서 Tom의 eng와 math 점수를 100으로 변경
# df에서 Jim, Tim의 kor, gym 점수를 null 값으로 변경
# df에서 math와 gym 점수를 바꿔서 출력

df.loc['Sam','math'] = 100
df.loc['Tom',['eng','math']] =100
df.loc[['Jim','Tim'],['kor','gym']] = np.nan
df['asd'] = df.iloc[:,3]
df.iloc[:,3] = df.loc[:,'gym']
df.iloc[:,4] = df['asd']
df = df.drop('asd',axis=1)
df
```

```python
# [과제] mtcars.csv를 불러와서 0~1 스케일로 정규화한 후
# qsec 컬럼이 0.5보다 큰 관측치의 개수를 구하시오.

from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()

mt = pd.read_csv('./dataset/mtcars.csv',index_col=0)
mts = mm.fit_transform(mt)
mt1 = pd.DataFrame(mts,columns = mt.columns)
len(mt1[mt1.qsec>0.5])
```

```python
# [연습문제] data = np.random.randn(10000).reshape(100,100)에
# I1,2,3... C1,2,3...의 인덱스와 컬럼명을 붙여서 데이터프레임화 시켜보세요.
data = np.random.randn(10000).reshape(100,100)
index = []
col = []
for i in range(1,101):
    index.append('I'+str(i))
for c in range(1,101):
    col.append('C'+str(c))
df = pd.DataFrame(data,index,columns=col)
df
```
