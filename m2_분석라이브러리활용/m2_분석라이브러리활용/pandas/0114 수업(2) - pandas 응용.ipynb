### 그룹 연산
- 복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나누어 관찰할 수 있으며,<br>
    이런 방식으로 분할 처리하는 것을 '그룹 연산'이라 함
- 그룹 연산은 데이터를 집계, 변환, 필터링하는 데 효율적이며<br>
    판다스 groupby() 메소드를 사용함
- 그룹 객체 만들기(분할)
- 그룹 연산 메소드(적용, 결합)

```python
titanic = sns.load_dataset('titanic')
df=titanic.loc[:,['age','sex','class','fare','survived']]
print('승객 수:',len(df))
display(df.head())
```

### 적용-결합
<br>

#### 데이터 집계(aggrigation)
- 집계 연산을 처리하는 사용자 정의 함수를 그룹 객체에 적용하려면<br>
    agg() 메소드 사용
    
- 모든 열에 여러 함수를 매핑:<br>
    group객체.agg([함수1,함수2,함수3,...]) (함수의 '리스트')
    
- 각 열마다 다른 함수를 매핑:<br>
    group객체.agg({'열1':함수1,'열2':함수2,...}) (열과 함수의 '딕셔너리)
    
<br>

#### 변환(transform)
- 원소의 본래 행 인덱스와 열 이름을 기준으로 연산 결과를 반환
- 데이터 변환 연산: group객체.transform(매핑 함수)

<br>

#### 필터링(filter)
- 그룹 객체에 filter() 메소드를 적용할 때 조건식을 가진 함수를 전달하면<br>
    조건이 참인 그룹만을 남김
- 그룹 객체 필터링:<br>
    group객체.filter(조건 함수식)
    
<br>

#### 객체에 함수 매핑(apply)
- 판다스 객체의 개별 원소를 특정 함수에 일대일로 매핑<br>
    사용자가 원하는 대부분의 연산을 그룹 객체에 적용
- group객체.apply(매핑 함수)

```python
# [연습문제] 최댓값-최솟값을 반환하는 사용자 함수를 정의한 후
# agg() 함수의 인수로 넣어 그룹별로 집계하여
# agg_minmax 이름으로 저장 후 처음 5개행을 출력하세요.

def minmax(a):
    return a.max()-a.min()

df=titanic.loc[:,['age','sex','class','fare','survived']]
#df1 = df[['sex']].replace(['male','female'],[0,1])
#df[['sex']] = df1
#display(df.head())
gr = df.groupby(['class'])
agg_minmax = gr.agg(minmax)
agg_minmax
```

```python
# [연습문제] 자료가 평균 값에서 표준편차의 몇 배 떨어져 있는지를 평가하는 사용자 함수를 작성하세요.
# 표준편차는 1이고, 자료와 평균 사이의 거리가 2면 표준편차의 2배만큼 떨어져 있다.
# 자료와 평균 간의 차를 std(표준편차)로 나눠주면 된다.

# z점수라는 건데
# z점수 = (자료값-평균)/표준편차 (편차를 표준편차로 나눠줬다)

def z_score(a):
    return (a-a.mean())/a.std()
```

```python
gr[['age']].transform(z_score) # 이거 타입은 데이터프레임. []로 한 번 더 감싸줬기 때문에.
```

```python
# [연습문제] 자료가 평균 값에서 표준편차의 몇 배 떨어져 있는지를 평가하는 사용자 함수를 작성하세요.
# 표준편차는 1이고, 자료와 평균 사이의 거리가 2면 표준편차의 2배만큼 떨어져 있다.
# 자료와 평균 간의 차를 std(표준편차)로 나눠주면 된다.

def z_score(a):
    return (a-a.mean())/a.std()

# [연습문제] 위에서 구한 사용자 함수를 이용하여 age 칼럼을 transform() 함수로 변환하세요.
gr[['age']].transform(z_score)
```

```python
# [연습문제] 위에서 구한 사용자 함수를 이용하여 age 칼럼을 transform() 함수로 변환하세요.
gr.age.transform(z_score) # 이거 타입은 시리즈
```

```python
# [연습문제] 위에서 구한 사용자 함수를 이용하여 age 칼럼을 apply() 함수로 변환하세요.
display(gr.age.apply(z_score))
display(gr[['age']].apply(lambda x : z_score(x))) # age의 각 값이 x
```
