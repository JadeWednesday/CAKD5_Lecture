```python
import seaborn as sns
dftitan = sns.load_dataset('titanic')
dftitan
```

```python
dft = dftitan[dftitan.index<11].reset_index()
dft
```

```python
# 데이터 추출, 변경, 삭제, 정렬, 칼럼 병합 (sibsp+parch = family) 등을 수행하여
# 데이터에서 인사이트를 찾아보세요.

# 데이터 변경 - nan으로 출력되는 5번 인덱스 사람의 나이를 임의로 지정
np.random.seed(0)
dft.loc[5,'age'] = np.random.randint(10,80)
#dft.drop('index',axis=1) #inplace=True로 할 거면 한번 적용하고서 지워줘야 오류가 안 남
# 데이터 삭제 - pclass와 class는 비슷한 내용이므로 pclass 삭제, 데이터 집중을 위해 embarked, 데크 열 삭제
dft1 = dft.drop(['index','pclass','deck','embarked'],axis=1)

# 데이터 병합 - sibsp(형제와 배우자), parch(부모와 자식)란을 통틀어 family로 병합
dft2 = dft1.copy().drop(['sibsp','parch'],axis=1)
dft2.loc[:,'family'] = dft1.loc[:,'sibsp'] + dft1.loc[:,'parch']

# 데이터 정렬 - 나이와 요금 순으로 정렬(나이 우선)
dft3 = dft2.sort_values(by=['age','fare'],ascending=False).reset_index()
dft3.drop('index',axis=1,inplace=True)
dft3
```

```python
# 데이터 추출
print(dft3[['survived','sex']]) # 남성일 경우 사망, 여성일 경우 생존
# 익히 알고 있듯 여성과 어린아이를 먼저 구명보트에 태웠기 때문에 이런 결과가 도출
print(dft3[['class','fare']])
# 3등석에 비해 1등석은 적게는 2.5배, 많게는 10배까지 비쌌기 때문에 3등석 택하는 사람들이 많았을 것
print(dft3[['survived','family']])
# 가족, 형제 등 동반자가 있는 경우에 생존 확률이 더 높았다.
```

#### [과제] iloc, loc를 이용해서 아래 요구사항들을 출력하세요.
- df의 kor열을 series와 dataframe으로 각각 출력
- df에서 Sam의 math 점수를 출력 (2가지 방법)
- df에서 Tom의 eng와 sci를 출력 (2가지 방법)
- df에서 Jim, Tim의 eng, gym 점수를 출력
- df에서 Jim, Sam의 art, math, gym 점수를 출력

```python
# df의 kor열을 series와 dataframe으로 각각 출력
print(df.loc[:,'kor'], type(df.loc[:,'kor']),'\n')  # df['kor']
print(df.loc[:,:'kor'],type(df.loc[:,:'kor']),'\n') # df[['kor']]
# :'kor'도 되고, ['kor']도 된다. 이건 kor가 첫 번째 열이어서 가능했던 거니까
# 다음부턴 [[]]를 쓰도록 하자.

# df에서 Sam의 math 점수를 출력 (2가지 방법)
print(df.loc['Sam','math'],'\t')
print(df.iloc[1,3],'\n')

# df에서 Tom의 eng와 sci를 출력 (2가지 방법)
print(df.loc['Tom',['eng','sci']],'\n')
print(df.iloc[2,[1,5]],'\n')

# df에서 Jim, Tim의 eng, gym 점수를 출력
print(df.loc[['Jim','Tim'],['eng','gym']],'\n')

# df에서 Jim, Sam의 art, math, gym 점수를 출력
print(df.loc[:'Sam','art':'gym']) # 문자열일 땐 인덱스 범위 +1 이거 안 해줘도 됨
print(df.iloc[:2,2:5],'\n')       # 숫자일 때만 인덱스 범위 +1 해주고, 문자열은 직관적으로!
```
